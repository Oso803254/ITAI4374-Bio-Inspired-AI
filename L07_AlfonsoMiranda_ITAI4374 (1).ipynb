An LLM most closely resembles the **neocortical slow-learning system** from CLS theory. ## Experiment B4: Visualizing AI Attention (Code Exercise)

**Neuroscience concept:** Module 07 described attention as a "spotlight" that the brain shines on relevant information. In transformers, self-attention computes relevance scores between every pair of tokens — we can actually SEE where the model is "looking."

In this experiment, you will use a pre-trained transformer model to visualize its attention patterns.

### TODO B4: Run and Analyze the Attention Visualization


````python
# ── Experiment B4: Visualize Transformer Self-Attention ──
from transformers import AutoTokenizer, AutoModel
import torch

# Load a small pre-trained model (runs on CPU)
model_name = "bert-base-uncased"
print(f"Loading {model_name}... (this takes ~30 seconds on first run)")
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name, output_attentions=True)
model.eval()
print("✅ Model loaded!")
````

````python
# ── Visualize attention for a sentence ──
def visualize_attention(sentence, layer=11, head=0):
    """
    Shows where the model "looks" for each word in the sentence.
    
    - layer: which transformer layer to visualize (0-11, higher = more abstract)
    - head: which attention head (0-11, different heads attend to different things)
    """
    # Tokenize
    inputs = tokenizer(sentence, return_tensors="pt")
    tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0])
    
    # Get attention weights
    with torch.no_grad():
        outputs = model(**inputs)
    
    # Extract attention from specified layer and head
    attention = outputs.attentions[layer][0, head].numpy()
    
    # Plot
    fig, ax = plt.subplots(figsize=(10, 8))
    sns.heatmap(attention, xticklabels=tokens, yticklabels=tokens,
                cmap="Blues", ax=ax, vmin=0, vmax=0.5)
    ax.set_title(f"Attention Pattern — Layer {layer}, Head {head}\n"
                 f"Each row shows where that word \"looks\"\n"
                 f"Brighter = more attention",
                 fontsize=12)
    ax.set_xlabel("Attended TO (where the model looks)")
    ax.set_ylabel("Attended FROM (which word is looking)")
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

# ── Run it! ──
visualize_attention("The cat sat on the mat because it was tired.")
print("\n💡 Look at the row for 'it' — does the model attend to 'cat' or 'mat'?")
print("   This is the model resolving a pronoun reference — a form of attention!")
````

````python
# ── TODO B4: Try your own sentences ──
# Change the sentence below and run the cell to see different attention patterns.
# Try sentences where it's ambiguous what a word refers to.

# Sentence 1: Pronoun resolution
visualize_attention("The doctor told the nurse that she was late.", layer=11, head=0)

# Sentence 2: Try a different layer (early layers attend locally, late layers attend globally)
visualize_attention("The doctor told the nurse that she was late.", layer=1, head=0)
````

### ✅ B4: Observations

**What patterns did you notice in the attention heatmap for "The cat sat on the mat because it was tired"?**

In the Layer 11 heatmap, the row corresponding to the token "it" shows the strongest attention weight directed toward "cat" rather than "mat." This reflects BERT's contextual understanding of pronoun coreference: "it was tired" logically refers to the animate subject (the cat) rather than the inanimate object (the mat). The diagonal also shows moderate self-attention throughout — each word attends somewhat to itself — and the [CLS] and [SEP] special tokens receive high attention from nearly every word, which is characteristic of BERT's pooling behavior.

**How did the attention pattern differ between Layer 1 (early) and Layer 11 (late)?**

Layer 1 attention is predominantly **local**: tokens most strongly attend to immediately adjacent tokens and the sentence boundaries ([CLS], [SEP]). The pattern is relatively uniform and undifferentiated, reflecting early syntactic proximity rather than semantic reasoning. Layer 11 attention is significantly more **structured and long-range**: semantically related tokens that are far apart in the sentence (e.g., "it" → "cat") show high mutual attention, indicating the model has built up sufficient contextual representation to reason about non-local relationships. The late-layer heatmap shows clearly interpretable semantic clusters and cross-sentence reference patterns that are absent in the early layers.

**Neuroscience connection — Hierarchical attention:**

This early-to-late layer attention progression directly parallels the **visual processing hierarchy** described in Modules 05 and 07. In the visual cortex, V1 neurons respond to local features — oriented edges within a tiny receptive field (a few degrees of visual angle). As signals ascend through V2, V4, and into the inferotemporal cortex, receptive fields grow larger and representations become more abstract and integrative: faces, objects, semantic categories. The transformer mirrors this hierarchy: early layers (analogous to V1/V2) perform local, syntactic feature detection across nearby tokens; later layers (analogous to IT cortex) integrate over entire sentences to resolve semantic relationships like pronoun reference, subject-verb agreement, and coreference. This convergence is not coincidental — both the biological visual system and the transformer are hierarchical function approximators that build increasingly abstract representations from simple local features, a design principle that appears to be domain-general across natural and artificial intelligence systems.

### ✅ C2: Reflection Questions

**R1.** What was the biggest difference you observed between how the AI handles memory versus how the brain handles memory? Which Module 06 concept best explains this difference?

The most striking difference is the **binary nature of AI memory loss compared to the gradual, curve-shaped forgetting of the human brain**. Human forgetting follows the Ebbinghaus forgetting curve — exponential decay over time, modulated by sleep consolidation, emotional salience, retrieval practice, and interference. The AI, by contrast, experienced zero forgetting within a session (Experiment A2 confirmed perfect recall after 12 interfering exchanges) but total, instantaneous forgetting between sessions. This asymmetry — perfect short-term retention, no long-term consolidation — is best explained by Module 06's **Complementary Learning Systems (CLS) theory**. The CLS framework describes two cooperating systems: a fast hippocampal system for rapid encoding and a slow neocortical system for stable long-term storage. The AI has only an analog of the neocortical component (pre-trained weights) with no functional hippocampus. There is no fast-binding system to create new memories from experience, and no consolidation process to transfer session-level learning into persistent storage. The result is a system with vast crystallized knowledge but no capacity for episodic growth.

---

**R2.** What was the biggest difference you observed between how the AI handles attention versus how the brain handles attention? Which Module 07 concept best explains this difference?

The most significant difference is that the AI lacks a **pre-attentive automatic monitoring system** — the biological mechanism best illustrated by the cocktail party effect (Experiment B3). In the brain, attention operates on two tracks: a fast, automatic, pre-attentive track that continuously scans the environment for self-relevant or emotionally salient signals (driven by the alerting network and the amygdala), and a slower, deliberate, executive track that consciously directs focus. The AI has only the deliberate track — it processes all input tokens with equal initial access through self-attention and then responds according to trained objectives. It has no equivalent of the background monitoring system that would automatically flag "Jordan" as a high-priority signal separate from task-relevant processing. Module 07's **Posner three-network model** best explains this gap: the alerting network (sustained readiness via norepinephrine and locus coeruleus) is entirely absent in the transformer architecture. The AI's executive attention is surprisingly capable, but its alerting and automatic-salience systems are missing entirely.

**R3.** If you could add ONE brain-inspired feature to the AI system you tested, what would it be and why? Be specific — name the brain structure or mechanism that inspires your recommendation and explain how it would improve the AI.

I would add a functional analog to the **hippocampal fast-binding memory system** — specifically, a persistent episodic memory module that can encode specific conversational events as distinct retrievable traces and retain them across sessions. The biological hippocampus enables one-shot learning: a single meaningful experience can be encoded as a unique episodic memory indexed by context, time, and emotional valence, and later retrieved in full. This is what allows a human to remember a conversation from last Tuesday without it being rehearsed. The AI currently has no equivalent — everything resets at session close (Experiment A2 confirmed this limitation). Implementing a hippocampal-inspired module would mean the AI could remember that a user mentioned their dog is named Biscuit, that they are working on a data science project, or that they prefer concise answers — not just within a session, but across sessions, growing a persistent, user-specific episodic store. This would transform the AI from a stateless function (same weights every time) into a genuinely adaptive system that learns from its accumulated experience with each user. Combined with a consolidation mechanism (periodically compressing episodic traces into updated semantic knowledge, analogous to sleep-based hippocampal replay), the system could become meaningfully personalized over time in a way that current LLMs fundamentally cannot achieve.
